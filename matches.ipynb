```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_eng = pd.read_csv('./builtin/E.csv')
df_bel = pd.read_csv('./builtin/B1.csv')
df_frn = pd.read_csv('./builtin/F1.csv')
df_ity = pd.read_csv('./builtin/I1.csv')
df_ntd = pd.read_csv('./builtin/N1.csv')
df_ptg = pd.read_csv('./builtin/P1.csv')
df_tur = pd.read_csv('./builtin/T1.csv')
df_ger = pd.read_csv('./builtin/D1.csv')
df_spn = pd.read_csv('./builtin/SP1.csv')

```

```python
objs=[df_eng,df_bel,df_frn,df_ity,df_ntd,df_ptg,df_tur,df_ger,df_spn]
df=pd.concat(objs, axis=0, join='outer', ignore_index=False, keys=None)
```

```python
cols=df.iloc[:,0:27]
```

```python
cols
```

```python
cols.value_counts()
```

```python
cols.columns
```

```python

cols.dropna(inplace=True)
print(cols.isnull().sum())
```

```python
cols
```

```python
cols['FTR'] = cols['FTR'].map({'H': 1, 'D': 0, 'A': -1})
```

```python
cols
```

```python
features = cols[['FTHG', 'FTAG', 'HS', 'AS', 'HF', 'AF']]
target = cols['FTR']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize and train the model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions
predictions = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print(f'Model Accuracy: {accuracy:.2f}')
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion Matrix
cm = confusion_matrix(y_test, predictions)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Aggregate data by season and team
goals_trend = cols.groupby(['HomeTeam'])['FTHG'].sum()

# Plotting
plt.figure(figsize=(12, 6))
goals_trend.plot(kind='line', marker='o')
plt.title('Goals Scored by Teams Over Seasons')
plt.xlabel('Teams')
plt.ylabel('Goals Scored')
plt.xticks(rotation=45)
plt.legend(title='Teams')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Aggregate wins by team across all seasons
wins_summary = cols.groupby(['HomeTeam'])['FTR'].sum().sort_values()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=wins_summary.index, y=wins_summary.values, palette='viridis')
plt.title('Total Wins by Team')
plt.xlabel('Teams')
plt.ylabel('Total Wins')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create a pivot table for heatmap
heatmap_data =cols.pivot_table(index='HomeTeam', columns='AwayTeam', values='FTHG', aggfunc='sum')

# Plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='g')
plt.title('Heatmap of Goals Scored by Team Across Seasons')
plt.xlabel('Season')
plt.ylabel('Team')
plt.show()
```
